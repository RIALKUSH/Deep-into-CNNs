```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("train.csv")
train
```

```python
train.describe()
```

```python
test=pd.read_csv("test.csv")
test
```

```python
test.describe()
```

## Analyse "target"

```python
import matplotlib.pyplot as plt
train.groupby('target').target.count().plot.bar(ylim=0)
plt.show
```

```python
target=train.target
```

```python
train.drop('target',axis=1,inplace=True)
train
```

```python
import seaborn as sns
corr_matrix = train.corr().abs()
sns.heatmap(corr_matrix)
```

## Merge train and test datasets

```python
all_data = train.append(test)
all_data
```

```python
all_data.drop('id',axis=1,inplace=True)
all_data
```

### One hot encode target

```python
target=pd.get_dummies(target)
target
```

```python
#check for null values
all_data.isnull().sum().sum()
```

## Normalization

```python
all_data = (all_data.max()-all_data)/(all_data.max()-all_data.min())
all_data
```

## Define independant and dependant variables

```python
# separate train and test data
y = target
X = all_data[: len(train)]
X_test = all_data[len(train) :]
```

## Split training set

```python
#split train set for testing
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.10, random_state=1, stratify=y)
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape
```

## Model Selection

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn import linear_model

model = MultiOutputRegressor(linear_model.BayesianRidge(n_iter=10000)).fit(X_train, y_train)
model.score(X_train, y_train)
```

## Predict on validation set

```python
y_pred = model.predict(X_val)
model.score(X_val, y_val)
```

## Predict on test sdet

```python
predictions = model.predict(X_test)
predictions.shape
```

```python
predictions[predictions < 0] = 0
```

## Generate Submission File

```python
sub = pd.DataFrame({ 'id': test.id, 'Class_1': predictions[:, 0], 'Class_2': predictions[:, 1], 'Class_3': predictions[:, 2], 'Class_4': predictions[:, 3], 'Class_5': predictions[:, 4], 'Class_6': predictions[:, 5], 'Class_7': predictions[:, 6], 'Class_8': predictions[:, 7], 'Class_9': predictions[:, 8] })
sub.to_csv("submission_sklearn.csv", index=False)
```
