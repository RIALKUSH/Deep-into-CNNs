# Diabetes Prediction Using Logistic Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
```

## Loading Data

```python
data = pd.read_csv("diabetes2.csv", header=None, skiprows=[0], names=col_names)
```

```python
data
```

```python
y = data.label.values
x_data = data.drop(["label"],axis=1)
```

```python
x_data
```

## Normalization

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values
```

```python
x
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = None)

x_train = x_train.T
x_test = x_test.T
y_train = y_train.T
y_test = y_test.T
```

## Parameter Initialize and Sigmoid Function

```python
def initialize_weights_and_bias(dimension):
    w = np.full((dimension,1),0.01)
    b = 0.0
    return w,b

def sigmoid(z):
    y_head = 1 / (1+np.exp(-z))
    return y_head
```

## Forward and Backward Propagation

```python
def forward_backward_propagation(w,b,x_train,y_head):
    z = np.dot(w.T,x_train) + b
    y_head = sigmoid(z)
    loss = -y_train*np.log(y_head) - (1-y_train)*np.log(1-y_head)
    cost = (np.sum(loss)) / x_train.shape[1]
    #backward propogation
    derivative_weight = (np.dot(x_train,((y_head-y_train).T)))/x_train.shape[1]
    derivative_bias = np.sum(y_head-y_train)/x_train.shape[1]
    gradients = {"derivative_weight": derivative_weight,"derivative_bias": derivative_bias}
    return cost,gradients
```

## Prediction

```python
def predict(w,b,x_test):
    # x_test is a input for forward propagation
    z = sigmoid(np.dot(w.T,x_test)+b)
    Y_prediction = np.zeros((1,x_test.shape[1]))
    for i in range(z.shape[1]):
        if z[0,i]<= 0.5:
            Y_prediction[0,i] = 0
        else:
            Y_prediction[0,i] = 1

    return Y_prediction
```

## Updating (Learning) Parameters

```python
def update(w, b, x_train, y_train, learning_rate,number_of_iterarion):
    cost_list = []
    cost_list2 = []
    correctness_list = []
    index = []
    for i in range(number_of_iterarion):
        # make forward and backward propagation and find cost and gradients
        cost,gradients = forward_backward_propagation(w,b,x_train,y_train)
        cost_list.append(cost)
        
        dw= gradients["derivative_weight"]
        db= gradients["derivative_bias"]
        
        # lets update
        w = w - learning_rate * dw
        b = b - learning_rate * db
        
        Y_prediction = predict(w - learning_rate * dw,b - learning_rate * db,x_train)
        correctness = 100 - np.mean(np.abs(Y_prediction - y_train)) * 100
        
        correctness_list.append(correctness)
        cost_list2.append(cost)
        index.append(i)
        print ("Epoch %i: (Cost: %f, Correctness: %f)" %(i+1, cost, correctness))
        
    # we update(learn) parameters weights and bias
    parameters = {"weight": w,"bias": b}
    
    plt.plot(index,cost_list2)
    plt.xlabel("Epochs")
    plt.ylabel("Cost")
    plt.title("Cost after each Epoch")
    plt.show()
    
    plt.plot(index,correctness_list)
    plt.xlabel("Epochs")
    plt.ylabel("Correctness")
    plt.title("Correctness after each Epoch")
    plt.show()
    return parameters, gradients, cost_list
```

## Defining Logistic Regression Function

```python
def logistic_regression(x_train, y_train, x_test, y_test, learning_rate ,  num_iterations):
    dimension =  x_train.shape[0]
    w,b = initialize_weights_and_bias(dimension)
    
    parameters, gradients, cost_list = update(w, b, x_train, y_train, learning_rate,num_iterations)
    
    y_prediction_test = predict(parameters["weight"],parameters["bias"],x_test)
    
    print("test accuracy(Correct/Total): {}/{}".format(y_test.shape[0]-np.sum(np.abs(y_prediction_test - y_test)),(y_test.shape[0])))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
```

```python
logistic_regression(x_train, y_train, x_test, y_test,learning_rate = 0.01, num_iterations = 200)
```

```python
logistic_regression(x_train, y_train, x_test, y_test,learning_rate = 0.5, num_iterations = 200)
```

```python
logistic_regression(x_train, y_train, x_test, y_test,learning_rate = 2, num_iterations = 200)
```
